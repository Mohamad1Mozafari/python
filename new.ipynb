```python
import stanza
import spacy 
# Download Persian model (run once)
stanza.download('fa')

# Initialize the pipeline for Persian
nlp = stanza.Pipeline('fa')

# Your Persian text
text = "قیمت تلفن امروز 20 است فردا ارزان میشود "

# Process the text
doc = nlp(text)

# Extract keywords (nouns and proper nouns)
keywords = []
for sentence in doc.sentences:
    for word in sentence.words:
        if word.upos in ['NOUN', 'PROPN']:
            keywords.append(word.text)

# Remove duplicates
unique_keywords = list(set(keywords))

print("Keywords:", unique_keywords)
```

```python
from selenium import webdriver
from bs4 import BeautifulSoup
import time

# Setup browser
options = webdriver.ChromeOptions()
options.add_argument('--headless')
driver = webdriver.Chrome(options=options)

# Load page
driver.get("https://www.digikala.com/")
time.sleep(5)

soup = BeautifulSoup(driver.page_source, 'html.parser')

# Remove all <a> tags from a copy of soup
soup_no_links = soup.__copy__()
for a in soup_no_links.find_all('a'):
    a.decompose()  # Remove all <a> tags and their content

# Get remaining visible text
non_linked_text = soup_no_links.get_text(separator=' ', strip=True)
print("\n--- Non-linked Text ---\n")
print(non_linked_text)

linked_text = []
for a in soup.find_all('a', href=True):
    text = a.get_text(strip=True)
    if text:
        linked_text.append(text)

print("\n--- Linked Text ---\n")
print('\n'.join(linked_text))

driver.quit()
```

```python
cut=int(len(linked_text) * 0.05)
```

```python
from sentence_transformers import SentenceTransformer, util
model = SentenceTransformer('distiluse-base-multilingual-cased-v1')
words = linked_text[:cut]
keyword_text = " ".join(words)
keyword_embedding = model.encode(keyword_text, convert_to_tensor=True)
nlp = stanza.Pipeline('fa')
doc = nlp(keyword_text)
sentences = [" ".join([word.text for word in sentence.words]) for sentence in doc.sentences]

# Filter relevant sentences
SIMILARITY_THRESHOLD = 0.6
matched_sentences = []

for sent in sentences:
    sent_embedding = model.encode(sent, convert_to_tensor=True)
    similarity = util.pytorch_cos_sim(keyword_embedding, sent_embedding).item()
    if similarity >= SIMILARITY_THRESHOLD:
        matched_sentences.append((sent, similarity))

# --- Step 5: Print Top Matches ---
matched_sentences = sorted(matched_sentences, key=lambda x: x[1], reverse=True)

print("\nجملات مرتبط:")
for sent, score in matched_sentences:
    # print(f"{sent} [شباهت: {score:.2f}]")
    print(f"{sent}")
```

```python
import stanza
import trafilatura
from sentence_transformers import SentenceTransformer, util

# Load Stanza Persian model
# stanza.download('fa')
nlp = stanza.Pipeline('fa')

# Load a multilingual model that supports Persian
model = SentenceTransformer('distiluse-base-multilingual-cased-v1')

# --- Step 1: Persian Input Query ---
query = "قیمت خانه در تهران چقدر است؟"

# Process with Stanza to get noun-based keywords
doc = nlp(query)
keywords = []
for sentence in doc.sentences:
    for word in sentence.words:
        if word.upos in ['NOUN', 'PROPN']:
            keywords.append(word.text)

print("کلیدواژه‌ها:", keywords)

# --- Step 2: Embed the keywords as a single string ---
keyword_text = " ".join(keywords)
keyword_embedding = model.encode(keyword_text, convert_to_tensor=True)

# --- Step 3: Extract Persian Webpage Text ---
url = "https://www.maskannews.com/"  # use a Persian real estate or news site
html = trafilatura.fetch_url(url)
text = trafilatura.extract(html, no_fallback=True)

if not text:
    print("متن یافت نشد.")
    exit()

# --- Step 4: Split text into sentences and compute similarity ---
doc = nlp(text)
sentences = [" ".join([word.text for word in sentence.words]) for sentence in doc.sentences]

# Filter relevant sentences
SIMILARITY_THRESHOLD = 0.5
matched_sentences = []

for sent in sentences:
    sent_embedding = model.encode(sent, convert_to_tensor=True)
    similarity = util.pytorch_cos_sim(keyword_embedding, sent_embedding).item()
    if similarity >= SIMILARITY_THRESHOLD:
        matched_sentences.append((sent, similarity))

# --- Step 5: Print Top Matches ---
matched_sentences = sorted(matched_sentences, key=lambda x: x[1], reverse=True)

print("\nجملات مرتبط:")
for sent, score in matched_sentences:
    print(f"{sent} [شباهت: {score:.2f}]")
```
